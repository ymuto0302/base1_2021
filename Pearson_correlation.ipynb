# (事前知識) Pearson の相関計数
Pearson の相関係数は「2変数が連動して変化する傾向の度合い」を測定する指標である。

長さ n の２つの時系列データ $\boldsymbol{x} = (x_1, x_2, \cdots, x_n)$，$\boldsymbol{y} = (y_1, y_2, \cdots, y_n)$ の相関は次式で与えられる。

$\displaystyle r = \frac{ \sum_{i=1}^{n} ( x_{i} - \overline{x} ) ( y_{i} - \overline{y}) }{ \sqrt{ \sum_{i=1}^{n} ( x_{i} - \overline{x})^2 } \sqrt{ \sum_{i=1}^{n} ( y_{i} - \overline{y})^2 } }$

相関を求めるには，２つの Series を corr() で繋ぐ形式と DataFrame に対して corr() を適用する形式がある。

## ２つの Series を corr() で繋ぐパターン

```python
# データの定義
import pandas as pd

series_a =pd.Series([1, 2, 1, 3, 4, 1])
series_b = pd.Series([4, 5, 4, 6, 7, 5])
```

```python
# 可視化
import matplotlib.pyplot as plt

series_a.plot(label="A")
series_b.plot(label="B")
plt.legend()
plt.show()
```

```python
# 相関計数の算出
series_a.corr(series_b)
```

## DataFrame に対して corr() を適用するパターン

```python
import pandas as pd

# データの定義
df = pd.DataFrame({"A":[1, 2, 1, 3, 4, 1],
                   "B":[4, 5, 4, 6, 7, 5],
                   "C":[10, 9, 10, 8, 7, 10]})

# 可視化
df.plot()
```

```python
# 相関計数
df.corr()
```
